# 1. Introduction

Customer churn is a critical concern for businesses, especially in competitive industries such as telecommunications. Predicting customer churn effectively enables organizations to retain their valuable customers by implementing targeted interventions, such as offering personalized discounts or plans. This project aims to leverage machine learning techniques to build predictive models that help identify potential churners and provide insights into the factors influencing customer churn.

## 1.1 Business Description

The Telecom industry operates in a highly competitive market, where retaining high-value customers is essential to maintaining revenue and market share. Customers often switch to competitors due to better service offerings, price differences, or dissatisfaction. This case study focuses on a telecom operator aiming to reduce churn rates by predicting customer behavior and understanding the key variables driving churn decisions. By addressing these challenges, the business can take proactive measures to enhance customer satisfaction and loyalty.

## 1.2 Problem Statement

In this case study, the goal is to build machine learning models that can:
- Predict whether high-value customers will churn in the near future.
- Identify the important variables that strongly predict churn, offering insights into why customers leave.
  
The dataset contains features describing customer usage patterns, recharge behaviors, and interactions. The challenge involves creating models that balance interpretability and predictive accuracy while addressing issues such as data dimensionality and the presence of unstructured data.

## 1.3 Objective of the Project

The primary objectives of this project are:

- Prediction of Customer Churn: Develop a machine learning model to predict whether high-value customers are likely to churn. This information can guide the company in implementing strategies like special offers or discounts to retain these customers.
- Identification of Churn Indicators: Create models to identify the key features contributing to customer churn. These insights can help businesses improve their services and address root causes of dissatisfaction.
- Evaluation Using Multiple Metrics: Evaluate the models on metrics such as accuracy, precision, recall, and F1-score to ensure they align with business goals, such as focusing on customers with high churn probability.
Recommendation of Retention Strategies: Based on model insights, recommend actionable strategies to manage and reduce customer churn effectively.

```python
def plot_combined_boxplot_with_hue(df, selected_columns, hue_col='churn_probability'):
    """
    Plots a single box plot for multiple columns with hue for differentiation.
    
    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - selected_columns (list of str): List of columns to plot.
    - hue_col (str): The column to use as hue for grouping in the box plot.
    """
    # Melt the DataFrame to have columns in a single 'variable' column
    melted_df = df.melt(id_vars=hue_col, value_vars=selected_columns, 
                        var_name="Columns", value_name="Values")
    # Plot
    plt.figure(figsize=(14, 8))
    if hue_col:
        sns.boxplot(data=melted_df, x="Columns", y="Values", hue=hue_col, palette="pastel", fliersize=3)
    else:
        sns.boxplot(data=melted_df, x="Columns", y="Values", palette="flare", fliersize=3)
    
    plt.xticks(rotation=45, ha='right')
    plt.title("Box Plot of Selected Columns by Churn Probability")
    plt.xlabel("Columns")
    plt.ylabel("Values")
    plt.legend(title=hue_col, loc='upper right')
    plt.show()
```

```python
def evaluate_model(y_true, y_pred, model_name):
    print(f"Metrics for {model_name}:")
    print("Accuracy: %.4f" % accuracy_score(y_true, y_pred))
    print("Precision: %.4f" % precision_score(y_true, y_pred))
    print("Recall: %.4f" % recall_score(y_true, y_pred))
    print("F1 Score: %.4f" % f1_score(y_true, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("---" * 20)
```

```python
def train_and_evaluate_models(X_train, y_train, X_test, y_test):
    models = {
        'Logistic Regression': LogisticRegression(),
        'Random Forest': RandomForestClassifier(),
        'XGBoost': XGBClassifier(eval_metric='logloss'),  # Adding eval_metric to avoid warnings
        'GradientBoosting':GradientBoostingClassifier()
    }

    for model_name, model in models.items():
        print(f"\nTraining {model_name}...")
        model.fit(X_train, y_train)

        # Predictions for train and test sets
        pred_train = model.predict(X_train)
        pred_test = model.predict(X_test)

        # Evaluate on train set
        print(f"\nTrain Metrics for {model_name}:")
        evaluate_model(y_train, pred_train, model_name + " (Train)")

        # Evaluate on test set
        print(f"\nTest Metrics for {model_name}:")
        evaluate_model(y_test, pred_test, model_name + " (Test)")
```

### 5.1.1. Hyperparameter Tunning - Random Forest

```python
# Define the parameter grid
param_grid = {
    'max_depth': [2,3,5,10,20],
    'min_samples_leaf': [5,10,20,50,100,200],
    'n_estimators': [10, 25, 50, 100]
}

# Initialize the model
rf = RandomForestClassifier(random_state=42)

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,  # 5-fold cross-validation
    n_jobs=-1,
    verbose=2,
    scoring='accuracy'  # Adjust scoring metric based on the problem
)

# Fit the model on training data
grid_search.fit(df_train_pca, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Evaluate on test data
rf_model_pca = grid_search.best_estimator_
pred_train = rf_model_pca.predict(df_train_pca)
pred_test = rf_model_pca.predict(df_test_pca)

# Evaluate on train set
print(f"\nTrain Metrics for final_model:")
evaluate_model(y_train, pred_train, rf_model_pca)

# Evaluate on test set
print(f"\nTest Metrics for final_model:")
evaluate_model(y_test, pred_test, rf_model_pca)
```

### 5.1.2. Hyperparameter Tuning - XGBoost

```python
# Define the parameter grid for RandomizedSearchCV
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [3, 5, 7, 9, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3],
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
    'gamma': [0, 0.1, 0.2, 0.3, 0.4],
    'min_child_weight': [1, 3, 5, 7],
    'scale_pos_weight': [1, 5, 10, 15, 20]
}
# Initialize the XGBoost classifier
xgb_clf = XGBClassifier(n_jobs=-1)

# Use RandomizedSearchCV to search for the best parameters
random_search = RandomizedSearchCV(
    estimator=xgb_clf,
    param_distributions=param_grid,
    n_iter=100,  # Number of parameter settings sampled
    scoring='accuracy',  # Choose an evaluation metric that suits your problem
    cv=5,  # 5-fold cross-validation
    verbose=1,
    n_jobs=-1,
    random_state=42
)

# Fit the model
random_search.fit(df_train_pca, y_train)

# Print best parameters and best score
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)

# Evaluate on test data
xgb_model_pca = random_search.best_estimator_
y_pred = xgb_model_pca.predict(df_test_pca)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("\nTest Metrics for Tuned XGBoost:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Confusion Matrix:\n", conf_matrix)
```

```python
# Evaluate on test data
pred_train = xgb_model_pca.predict(df_train_pca)
pred_test = xgb_model_pca.predict(df_test_pca)
# Evaluate on train set
print(f"\nTrain Metrics for final_model:")
evaluate_model(y_train, pred_train, xgb_model_pca)

# Evaluate on test set
print(f"\nTest Metrics for final_model:")
evaluate_model(y_test, pred_test, xgb_model_pca)
```

```python
def refine_logistic_model(X_train, y_train, threshold_pval=0.05, threshold_vif=10):
    X = X_train.copy()  # Work with a copy of the training set to avoid modifying the original data
    columns_dropped = []  # Track dropped columns
    current_accuracy = None
    
    while True:
        # Add constant to X
        X_with_const = sm.add_constant(X)

        # Fit the GLM model
        model = sm.GLM(y_train, X_with_const, family=sm.families.Binomial()).fit()

        # Get p-values and VIFs
        pvalues = model.pvalues
        vif = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]

        # Create a DataFrame to analyze features
        stats = pd.DataFrame({
            'Feature': X_with_const.columns,
            'Pvalue': pvalues,
            'VIF': vif
        })

        # Drop the constant row for analysis
        stats = stats[stats['Feature'] != 'const']

        # Check if any p-value > threshold_pval exists
        if (stats['Pvalue'] > threshold_pval).any():
            # Find the feature with the highest p-value
            feature_to_drop = stats.loc[stats['Pvalue'].idxmax(), 'Feature']
            print(f"Dropping feature '{feature_to_drop}' due to high P-value ({stats['Pvalue'].max():.3f})")
        elif (stats['VIF'] > threshold_vif).any():
            # Find the feature with the highest VIF
            feature_to_drop = stats.loc[stats['VIF'].idxmax(), 'Feature']
            print(f"Dropping feature '{feature_to_drop}' due to high VIF ({stats['VIF'].max():.2f})")
        else:
            print("No features with high P-value or high VIF. Model refinement complete.")
            break

        # Drop the identified feature
        X.drop(columns=[feature_to_drop], inplace=True)
        columns_dropped.append(feature_to_drop)

    # Fit the final model and return results
    final_model = sm.GLM(y_train, sm.add_constant(X), family=sm.families.Binomial()).fit()
    final_accuracy = final_model.aic  # Example metric; adjust as needed
    print(f"Model refinement complete. Final model accuracy (AIC): {final_accuracy}")
    return final_model, X, columns_dropped
```
